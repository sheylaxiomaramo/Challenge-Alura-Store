### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Cálculo de ingresos por tienda
ingreso_tienda1 = tienda.sum()["Precio"]
ingreso_tienda2 = tienda2.sum()["Precio"]
ingreso_tienda3 = tienda3.sum()["Precio"]
ingreso_tienda4 = tienda4.sum()["Precio"]

# Mostrar resultados
print("Ingreso total tienda 1:", ingreso_tienda1)
print("Ingreso total tienda 2:", ingreso_tienda2)
print("Ingreso total tienda 3:", ingreso_tienda3)
print("Ingreso total tienda 4:", ingreso_tienda4)
```

# 2. Ventas por categoría

```python
# Cantidad de productos vendidos por categoría en cada tienda

categorias_tienda1 = tienda['Categoría del Producto'].value_counts()
categorias_tienda2 = tienda2['Categoría del Producto'].value_counts()
categorias_tienda3 = tienda3['Categoría del Producto'].value_counts()
categorias_tienda4 = tienda4['Categoría del Producto'].value_counts()

# Mostrar resultados
print("Categorías más populares - Tienda 1")
print(categorias_tienda1)
print("\nCategorías más populares - Tienda 2")
print(categorias_tienda2)
print("\nCategorías más populares - Tienda 3")
print(categorias_tienda3)
print("\nCategorías más populares - Tienda 4")
print(categorias_tienda4)
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio de cada tienda
calif_prom_tienda1 = tienda['Calificación'].mean()
calif_prom_tienda2 = tienda2['Calificación'].mean()
calif_prom_tienda3 = tienda3['Calificación'].mean()
calif_prom_tienda4 = tienda4['Calificación'].mean()

# Mostrar resultados
print(f"Calificación promedio Tienda 1: {calif_prom_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {calif_prom_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {calif_prom_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {calif_prom_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Contar ventas por producto en cada tienda
ventas_prod_t1 = tienda['Producto'].value_counts()
ventas_prod_t2 = tienda2['Producto'].value_counts()
ventas_prod_t3 = tienda3['Producto'].value_counts()
ventas_prod_t4 = tienda4['Producto'].value_counts()
```

```python
print("Tienda 1 - Producto más vendido:\n", ventas_prod_t1.head(1))
print("Tienda 1 - Producto menos vendido:\n", ventas_prod_t1.tail(1))

print("\nTienda 2 - Producto más vendido:\n", ventas_prod_t2.head(1))
print("Tienda 2 - Producto menos vendido:\n", ventas_prod_t2.tail(1))

print("\nTienda 3 - Producto más vendido:\n", ventas_prod_t3.head(1))
print("Tienda 3 - Producto menos vendido:\n", ventas_prod_t3.tail(1))

print("\nTienda 4 - Producto más vendido:\n", ventas_prod_t4.head(1))
print("Tienda 4 - Producto menos vendido:\n", ventas_prod_t4.tail(1))
```

```python
import matplotlib.pyplot as plt

# Función para graficar los top productos por tienda
def graficar_top_ventas(ventas, nombre_tienda):
    plt.figure(figsize=(8,4))
    ventas.head(5).plot(kind='bar', color='skyblue')
    plt.title(f'{nombre_tienda} - Top 5 Productos Más Vendidos')
    plt.ylabel('Cantidad Vendida')
    plt.xlabel('Producto')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Graficar para cada tienda
graficar_top_ventas(ventas_prod_t1, "Tienda 1")
graficar_top_ventas(ventas_prod_t2, "Tienda 2")
graficar_top_ventas(ventas_prod_t3, "Tienda 3")
graficar_top_ventas(ventas_prod_t4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio por tienda
envio_prom_t1 = tienda["Costo de envío"].mean()
envio_prom_t2 = tienda2["Costo de envío"].mean()
envio_prom_t3 = tienda3["Costo de envío"].mean()
envio_prom_t4 = tienda4["Costo de envío"].mean()

# Mostrar resultados
print(f"Costo de envío promedio Tienda 1: {envio_prom_t1:.2f}")
print(f"Costo de envío promedio Tienda 2: {envio_prom_t2:.2f}")
print(f"Costo de envío promedio Tienda 3: {envio_prom_t3:.2f}")
print(f"Costo de envío promedio Tienda 4: {envio_prom_t4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Diccionario con ingresos por tienda (asegúrate de tener estas variables calculadas previamente)
ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

# Gráfico de barras
plt.figure(figsize=(6,4))
plt.bar(ingresos.keys(), ingresos.values(), color='skyblue')
plt.title("Ingreso total por tienda")
plt.ylabel("Suma de precios vendidos (Ingreso)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Distribución de categorías en Tienda 1
cat_tienda1 = tienda['Categoría del Producto'].value_counts()

# Gráfico de pastel
plt.figure(figsize=(6,6))
plt.pie(cat_tienda1, labels=cat_tienda1.index, autopct='%1.1f%%', startangle=90)
plt.title("Distribución de productos por categoría - Tienda 1")
plt.axis('equal')  # Círculo
plt.show()
```

```python
# Gráfico de dispersión de Precio vs Calificación
plt.figure(figsize=(6,4))
plt.scatter(tienda2['Precio'], tienda2['Calificación'], alpha=0.6, color='coral')
plt.title("Relación entre Precio y Calificación - Tienda 2")
plt.xlabel("Precio")
plt.ylabel("Calificación")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
